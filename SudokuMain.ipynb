```python
currentPuzzle = tridagonT1N1
possibilities = BacktrackingSearch.solve_sudoku(currentPuzzle)
temp = Sudoku.Puzzle(currentPuzzle)
startTime = time.perf_counter()
Sudoku.solve(temp)
print(f"Elapsed Time", time.perf_counter() - startTime, "\n")
if len(possibilities) == 1:
    correctSoln = possibilities[0]
    startTime = time.perf_counter()
    Sudoku.solve(temp)
    print(f"Elapsed Time", time.perf_counter() - startTime, "\n")
    if temp.solved():
        if not BacktrackingSearch.compare_puzzles(temp.stringRep(), BacktrackingSearch.stringRepX(correctSoln)):
            print("Correct solution!\n")
            temp.display()
            print()
            BacktrackingSearch.printSoln(correctSoln)
    elif not temp.solved():
        if temp.compare(possibilities[0]):
            print("The bot is on the right track.")
            BacktrackingSearch.printSoln(correctSoln)
        else:
            print("The bot made a mistake")
            temp.display()
            print()
            BacktrackingSearch.printSoln(correctSoln)
elif len(possibilities) > 1:
    print("Caution: mutliple solutions for this puzzle!")
    print(f"Total Possibilities: {len(possibilities)}\n")
    for x in possibilities:
        BacktrackingSearch.printSoln(x)
        print()
```
